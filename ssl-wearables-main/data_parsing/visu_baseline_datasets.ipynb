```python
import numpy as np
import os
from labels import *
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def load_data(data_root):
    x_path =  os.path.join(data_root, 'X.npy')
    y_path =  os.path.join(data_root, 'Y.npy')
    pid_path =  os.path.join(data_root, 'pid.npy')
    X = np.load(x_path)
    y = np.load(y_path)
    pid = np.load(pid_path)
    print('X:shape ', X.shape)
    print('y:shape ', y.shape)
    print('pid:shape ', pid.shape)
    return X, y, pid
```

# 1. PAMAP2

### 1.1 visu for ten sec

```python
data_root = '/data/UKBB/pamap_100hz_w10_o5/'
x, y, pid = load_data(data_root)
```

### 1.2 Visu for each activity

```python
plt.rcParams["figure.figsize"] = (15,3)
```

```python
def visu_act(sample_key, label_dicts, x, my_y, mode='10s'):
    sample_label = pamap2_labels[sample_key]
    act_filter =  np.array(my_y == sample_key)
    act_X = x[act_filter]

    num_sample = 2
    if len(act_X) == 0:
        return
    sample_idx = np.random.choice(len(act_X), num_sample)

    samples = act_X[sample_idx]

    fig, axs = plt.subplots(1, 2)

    for i in range(len(samples)):
        sample = samples[i]    
        axs[i].plot(sample)
        axs[i].set_title(sample_label+ ' ' + mode)
```

```python
for key, value in pamap2_labels.items():
    visu_act(key, pamap2_labels, x, y)
```
