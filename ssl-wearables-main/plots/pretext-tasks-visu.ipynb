```python
import numpy as np
from sslearning.data.data_transformation import *
import matplotlib.pyplot as plt
import random

%load_ext autoreload
%autoreload 2
```

```python
sample_path = '/data/UKBB/SSL/mini_sample/5300274_90001_0_0.npy'
data = np.load(sample_path)
data.shape
```

```python
# Transformation 
sample = data[900]
sample.shape
```

# Original shape

```python
myplot = plt.plot(sample.T)
```

## 1. Rotation



```python
fig = plt.figure(figsize=(15,4))
for ii in range(8):
    ax = fig.add_subplot(2,4,ii+1)
    if ii == 0:
        ax.plot(sample.T)
    else:
        axis = np.random.uniform(low=-1, high=1, size=sample.shape[0])
        angle = np.random.uniform(low=-np.pi, high=np.pi)
        ax.plot(rotation(sample, axis, angle).T)
    ax.set_xlim([0,900])
    ax.set_ylim([-1.5,1.5])
```

## 2. Axis switch



```python
fig = plt.figure(figsize=(15,4))
for ii in range(7):
    ax = fig.add_subplot(2,4,ii+1)
    ax.plot(switch_axis(sample, ii).T)
    ax.set_xlim([0,900])
    ax.set_ylim([-1.5,1.5])
```

## 3. Flip
Flip an epoch on the temporal scale



```python
fig = plt.figure(figsize=(20,8))
for ii in range(2):
    ax = fig.add_subplot(2,4,ii+1)
    ax.plot(flip(sample, ii).T)
    ax.set_xlim([0,900])
    ax.set_ylim([-1.5,1.5])
```

## 4. permute
Distort an epoch by dividing up the sample into several segments and then permute them



```python
fig = plt.figure(figsize=(15,4))
for ii in range(8):
    ax = fig.add_subplot(2,4,ii+1)
    if ii == 0:
        ax.plot(sample.T)
    else:
        ax.plot(permute(sample, 1).T)
    ax.set_xlim([0,900])
    ax.set_ylim([-1.5,1.5])
```

## 5. Jitter
Add gaussian noise



```python
fig = plt.figure(figsize=(15,4))
for ii in range(8):
    ax = fig.add_subplot(2,4,ii+1)
    if ii == 0:
        ax.plot(sample.T)
    else:
        ax.plot(jitter(sample, 0.1).T)
    ax.set_xlim([0,900])
    ax.set_ylim([-1.5,1.5])
```
